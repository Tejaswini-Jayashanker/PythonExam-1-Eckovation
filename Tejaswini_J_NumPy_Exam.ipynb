```python
import numpy as np
```

```python
'''
Q1. Load the file data.npy using the numpy load function. Let this be called data. (1 marks)
Information
• data is a 2D float array whose second dimension is 4.
• Each row in data represents a unique person.
• The columns correspond to attributes/qualities of that person.
• They are age(yrs), height(m), weight(kg) and gender (0 for Male, 1 for female)
'''
```

```python
#One method to solve the above Question:
data = np.load("data.npy")
#Print the numpy array
#Col_0 --> Age
#Col_1 --> Height
#Col_2 --> Weight
#Col_3 --> MALE(0)/FEMALE(1)
print(data)
```

```python
'''
 Q2. Each of the following questions has 2 marks for a total of 10 Marks.
1. Sort the array according to age, and save it as data_age_sorted.npy.
2. Sort the male rows by height and save it as data_males_height_sorted.npy. 
3.Print the difference in mean values for age, height and width between males and females. 
4.Print how the top 25% tallest people are divided among males and females (in terms of
percentages).
5.Find the indices of the people who are in the heaviest 50 people as well as in the tallest 50 people.
Save it as big_people_idcs.npy
'''
```

```python
#One Method that serves as a solution to Question 2.

#PART - 2(A) sort the array according to age
sort_age_ind = np.argsort(data[:, 0])
sorted_arr_asper_age = data[sort_age_ind]
print(sorted_arr_asper_age)
np.save("data_age_sorted.npy", sorted_arr_asper_age)
```

```python
#PART - 2(B) Sort Male's height

#Sort the Height
height_sort_ind = np.argsort(data[:, 1])
height_sorted_arr = data[height_sort_ind]
print("HEIGHT_SORTED ARRAY:\r\n",height_sorted_arr)

#Sort as per Gender
male_height_sorted_arr = height_sorted_arr[ np.where( height_sorted_arr[:,3] == 0.0 ) ]
print("MALE HEIGHT SORTED ARRAY:\r\n",male_height_sorted_arr)
np.save("data_males_height_sorted.npy", male_height_sorted_arr)
```

```python
#SELF UNDERSTANDING
female_height_sorted_arr = height_sorted_arr[ np.where( height_sorted_arr[:,3] == 1.0 ) ]
print("FEMALE HEIGHT SORTED ARRAY:\r\n",female_height_sorted_arr)
```

```python
#Part - 2(C) Mean Differences
male_arr = data[ np.where(data[:, 3] == 0.0 ) ]
female_arr = data[ np.where(data[:, 3] == 1.0 ) ]
# print("MALE ARRAY:\r\n",male_arr)
# print("FEMALE ARRAY:\r\n",female_arr)
print("*"*50)
#MALE
means_male = np.mean(male_arr,axis = 0)
age_mean_male = means_male[0]
height_mean_male = means_male[1]
weight_mean_male = means_male[2]
print("Male Height Mean: ",height_mean_male)
print("Male Age Mean: ", age_mean_male)
print("Male Weight Mean: ",weight_mean_male)
print("*"*50)

#FEMALE
means_female = np.mean(female_arr,axis = 0)
age_mean_female = means_female[0]
height_mean_female = means_female[1]
weight_mean_female = means_female[2]
print("Female Height Mean: ",height_mean_female)
print("Female Age Mean: ", age_mean_female)
print("Female Weight Mean: ",weight_mean_female)
print("*"*50)

#Mean Differences
print( "Average Height Difference between Male and Female: ",abs(height_mean_male - height_mean_female) )
print( "Average Weight Difference between Male and Female: ", abs(weight_mean_male - weight_mean_female) )
print( "Average Age Difference between Male and Female: ", abs(age_mean_male - age_mean_female))
```

```python
'''
Solution for 2(D):
Tallest 25% Ratio

This cannot be divided as Male and Female are not of the same height in any data.
2.13m is the tallest and Males will be 100% and Females will be 0%
Thus taking 25 tallest heights and checking.
'''
ind = np.argsort(data[:,1])
h_arr = ind[::-1][:]
tall_short = data[h_arr]
top25 = tall_short[0:25, :]
print(top25)
```

```python
#Get Unique Heights in the data
hght_uniq_list = list(np.unique(data[:,1]))
hght_uniq_list.sort(reverse = True) #Tallest to shortest order

top25_heights = hght_uniq_list[:25] #Gives top 25 tallest heights in the data
print(top25_heights)
```

```python
#To check how many females and belong to each height category
unique,counts = np.unique(male_arr[:,1], return_counts=True)
dict_heights_counts = dict(zip(unique, counts))
```

```python
top_heights = {}
for height in top25_heights:
    for key in dict_heights_counts.keys():
        if height == key:
            top_heights[key] = dict_heights_counts[key]
print(top_heights)
```

```python
unique,counts = np.unique(female_arr[:,1], return_counts=True)
female_dict_heights_counts = dict(zip(unique, counts))
female_top_heights = {}
for height in top25_heights:
    for key in female_dict_heights_counts.keys():
        if height == key:
            female_top_heights[key] = female_dict_heights_counts[key]
print(female_top_heights)
```

```python
for male_key in top_heights.keys():
    for female_key in female_top_heights.keys():
            total = int( top_heights[male_key] + female_top_heights[female_key] )
            male_per = int(top_heights[male_key])
            female_per = int(female_top_heights[female_key])
            print("Male Percemtage for",male_key,": ", (male_per/total)*100)
            print("Female Percemtage for",female_key,": ", (female_per/total)*100)
```

```python
'''
Solution for top 25%
'''
sort_array = np.array(sorted(data, key = lambda y : -y[1]))
tot_rec = sort_array.shape[0] // 4 +1
sort_array = sort_array[0: tot_rec, :]
male_count = np.count_nonzero(sort_array[:,3] == 0.0 )
female_count = np.count_nonzero(sort_array[:,3] == 1.0 )
print("Male Percemtage for: ", (male_count/tot_rec)*100)
print("Female Percemtage for: ", (female_count/tot_rec)*100)
```

```python
# Solution for 2(E):

#Get Unique Heights in the data
ind = np.argsort(data[:,1])
h_arr = ind[::-1][:]
tall_short = data[h_arr]
top50_height = tall_short[0:50, :]
print(top50_height) #50 tallest ppl
```

```python
#Get Unique Heights in the data
ind = np.argsort(data[:,2])
h_arr = ind[::-1][:]
heavy_light = data[h_arr]
top50_weights = heavy_light[0:50, :]
print(top50_weights) #50 heaviest ppl
```

```python
common_array = np.array([x for x in set(tuple(x) for x in top50_height) & set(tuple(x) for x in top50_weights)])
```

```python
np.save("big_people_idcs.npy", common_array)
```
